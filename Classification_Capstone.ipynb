```python
import pandas as pd
import re
from langchain_community.llms import Replicate
import os
from google.colab import userdata

# ============== SETUP =================
api_token = userdata.get("REPLICATE_API_TOKEN")
os.environ["REPLICATE_API_TOKEN"] = api_token

# Load model
output = Replicate(
    model="ibm-granite/granite-3.3-8b-instruct",
    replicate_api_token=api_token,
)

# Set model parameters
parameters = {
 "top_k": 2,
 "top_p": 0,
 "max_tokens": 256,
 "min_tokens": 0,
}

# Point to the dataset file
csv_path = os.path.join(path, "Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv")

# Load dataframe
df = pd.read_csv(csv_path)

# Take first 50 reviews (modify in head x)
customer_reviews = df["reviews.text"].dropna().head(50).tolist()


# ============== CHUNKING FUNCTION =================
def chunk_list(lst, chunk_size):
    for i in range(0, len(lst), chunk_size):
        yield lst[i:i + chunk_size]


# ============== PARSING FUNCTION =================
def parse_response(response_text, customer_reviews):
    """
    Parse the model's response into structured rows with categories expanded.
    """
    rows = []
    # Clean up indents/tabs
    cleaned = re.sub(r"\s+\n", "\n", response_text.strip())
    cleaned = re.sub(r"\t", " ", cleaned)

    # Regex for block (now allows Unclear too)
    pattern = (
        r"Review (\d+):\s*Sentiment:\s*(Positive|Negative|Mixed|Unclear)"
        r"(?:\s*-\s*Positive Categories:\s*(.*))?"
        r"(?:\s*-\s*Negative Categories:\s*(.*))?"
    )

    matches = re.findall(pattern, cleaned, re.IGNORECASE)

    # Categories we want as columns
    categories = [
        "Battery Life", "Battery Quality", "Packaging", "Value for Money",
        "Shipping/Delivery", "Brand Satisfaction"
    ]

    for match in matches:
        review_id, sentiment, pos, neg = match
        pos = pos if pos else "none"
        neg = neg if neg else "none"

        # Normalize categories into dict {cat: Positive/Negative/None}
        cat_map = {c: "None" for c in categories}

        pos_cats = [c.strip() for c in pos.split(",") if c.strip().lower() != "none"]
        neg_cats = [c.strip() for c in neg.split(",") if c.strip().lower() != "none"]

        for c in categories:
            if any(c.lower() in pc.lower() for pc in pos_cats):
                cat_map[c] = "Positive"
            elif any(c.lower() in nc.lower() for nc in neg_cats):
                cat_map[c] = "Negative"

        # Add row
        rows.append({
            "ReviewID": int(review_id),
            "ReviewText": customer_reviews[int(review_id) - 1],  # <— Align with full list, not just batch
            "Sentiment": sentiment.strip(),
            **cat_map
        })

    return rows


# ============== RUN CLASSIFICATION =================
all_rows = []

for batch_num, review_batch in enumerate(chunk_list(customer_reviews, 5), start=1):
    start_idx = (batch_num - 1) * 5

    reviews_text = "\n".join(
        [f"Review {start_idx + i + 1}: {review}" for i, review in enumerate(review_batch)]
    )

    prompt = f"""
    Classify these reviews as sentiments: Positive, Negative, Mixed, or Unclear.

    Then, assign one or more of the following categories:
    Battery Life: Whether or not the battery lasts for a long time.
    Battery Quality: Whether or not the battery works well, consistent, enough power, and can be used on certain devices.
    Packaging: Whether or not the item arrived in good quality or condition.
    Value for Money: Price compared to quality/quantity.
    Shipping/Delivery: Whether the item arrived fast or late.
    Brand Satisfaction: Whether they mention Amazon Basics brand, compare to other brands, or want to order/buy again.

    For "Mixed" sentiment reviews, separate Positive vs Negative categories.
    For "Unclear", set both Positive Categories and Negative Categories as "none".

    IMPORTANT:
    Always output both lines:
    - Positive Categories: [list or none]
    - Negative Categories: [list or none]

    [OUTPUT FORMAT]
    Review [X]:
    Sentiment: [Positive / Negative / Mixed / Unclear]
    - Positive Categories: [list or none]
    - Negative Categories: [list or none]

    {reviews_text}
    """

    response = output.invoke(prompt, parameters)
    print(f"\n===== Batch {batch_num} =====")
    print(response)

    # Fix: pass the FULL `customer_reviews`, not just batch
    rows = parse_response(response, customer_reviews)
    all_rows.extend(rows)

# ============== SAVE CLEAN CSV =================
df_out = pd.DataFrame(all_rows)

# ====== MAKE PIVOT-READY DATA ======
categories = [
    "Battery Life", "Battery Quality", "Packaging", "Value for Money",
    "Shipping/Delivery", "Brand Satisfaction"
]

df_long = df_out.melt(
    id_vars=["ReviewID", "ReviewText", "Sentiment"],
    value_vars=categories,
    var_name="Category",
    value_name="Status"
)

# Sentiment summary with percentages
total_reviews = len(df_out)
sentiment_counts = df_out["Sentiment"].value_counts().reset_index()
sentiment_counts.columns = ["Sentiment", "Count"]
sentiment_counts["Percentage"] = (sentiment_counts["Count"] / total_reviews * 100).round(2)

# Add total row
summary_row = pd.DataFrame([{
    "Sentiment": "Total Reviews",
    "Count": total_reviews,
    "Percentage": 100.0
}])
sentiment_summary = pd.concat([sentiment_counts, summary_row], ignore_index=True)

print("\n=== Overall Sentiment Counts & Percentages ===")
print(sentiment_summary)

# Remove "None" clutter
df_long_no_none = df_long[df_long["Status"] != "None"]

# Category-level summary
summary_by_category = (
    df_long_no_none
    .groupby(["Category", "Status"])
    .size()
    .reset_index(name="Count")
)

# Add percentages (relative to total reviews, not total mentions)
summary_by_category["Percentage"] = (
    summary_by_category["Count"] / total_reviews * 100
).round(2)

print("\n=== Category Breakdown with Percentages (no None) ===")
print(summary_by_category)


# Export everything to one Excel with multiple sheets
with pd.ExcelWriter("granite_classification.xlsx", engine="xlsxwriter") as writer:
    df_out.to_excel(writer, sheet_name="Per_Review", index=False)  # clean classification
    df_long_no_none.to_excel(writer, sheet_name="Pivot_Ready", index=False)  # pivot version
    sentiment_summary.to_excel(writer, sheet_name="Sentiment_Summary", index=False)  # summary
    summary_by_category.to_excel(writer, sheet_name="Category_Breakdown", index=False)

print("\nSaved all outputs into granite_classification.xlsx (multiple sheets)")
```
